```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print('✅ Libraries loaded!')
```

```python
from sklearn.datasets import load_iris
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df['species'] = df['target'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
print(df.head())
```

```python
print('Shape:', df.shape)
print('\nMissing Values:\n', df.isnull().sum())
print('\nDuplicate Rows:', df.duplicated().sum())
print('\nBasic Stats:\n', df.describe())
```

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for i, col in enumerate(data.feature_names):
    sns.histplot(df[col], kde=True, ax=axes[i], color='steelblue')
    axes[i].set_title(col.split()[0])
plt.suptitle('Feature Distributions', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df.iloc[:, :4].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.countplot(x='species', data=df, palette='Set2')
plt.title('Class Distribution')
plt.tight_layout()
plt.show()
```

## ✅ Key Takeaways
- Dataset: 150 rows, 4 features, 3 classes
- No missing values found
- Petal length & width are highly correlated
- Always explore data before building any model!
